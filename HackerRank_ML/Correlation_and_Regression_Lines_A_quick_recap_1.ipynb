Here are the test scores of 10 students in physics and history:

Physics Scores:  15  12  8   8   7   7   7   6   5   3

History Scores:  10  25  17  11  13  17  20  13  9   15

Compute Karl Pearson’s coefficient of correlation between these scores.
Compute the answer correct to three decimal places.



Another way of computing the Pearson Corr. coeff


    

```python
import math
```

```python
Physics_Scores = [15 , 12 , 8 ,  8 ,  7 ,  7 ,  7 ,  6  , 5 ,  3]
History_Scores = [10 , 25,  17  ,11 , 13  ,17 , 20 , 13 , 9  , 15]
```

```python
def sum_of_sq(List):
    
    return sum([x**2 for x in List])
```

```python
def PPMC(X,Y):
    
    N = len(X)
    sum_XY = sum([a*b for a,b in zip(X,Y)])
    sum_X_sq = sum_of_sq(X)
    sum_Y_sq = sum_of_sq(Y)
    
    r = (N*sum_XY - sum(X)*sum(Y)) / math.sqrt ((N*sum_X_sq - sum(X)**2)*(N*sum_Y_sq - sum(Y)**2)) 
    
    #print("%.2f" %r)
    
    return r
```

```python
PPMC(Physics_Scores,History_Scores)
```

# Alternative calculation

Lets programm another formula to compute Pearson corr.coeff.


```python
def mean(List):
    
    return float(sum(List))/len(List)
```

```python
def var(List):
    
    var = 0
    
    var = sum( [(j - mean(List))**2  for j in List] )

    #for j in range(len(List)):
        
        #var += (List[j] - mean(List)) **2 
    
    return var
```

```python
def covariance(X,Y):
    
    covar = 0
    
    for j in range(len(X)):
        
        covar +=  (X[j] - mean(X)) * (Y[j] - mean(Y))
    
    return covar
```

```python
def Pearson_corr_coeff(X,Y):
    
    r = covariance(X,Y) / math.sqrt(var(X)*var(Y))
    
    return r
```

```python
Pearson_corr_coeff(Physics_Scores,History_Scores)
```
